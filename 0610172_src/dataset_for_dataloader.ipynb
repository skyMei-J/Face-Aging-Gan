```python
import pandas as pd
import os, sys
path = "/data/thumbnails128x128/"
output_path='../DATALOADER/'
dataset = pd.read_csv("train_label.txt",sep = "	",header = None)
dataset = dataset.values
for i in range(len(dataset)):
    print(dataset[i][0])
```

```python
import shutil

dirs = os.listdir( path )
dirs.sort()
count = 0
if not os.path.exists(output_path):
    os.makedirs(output_path)
if not os.path.exists(output_path+'train/'):
    os.makedirs(output_path+'train/')
if not os.path.exists(output_path+'test/'):
    os.makedirs(output_path+'test/')

    

    
for file in dirs:
    if file[-3:] !='txt':
        imglist = os.listdir(path+file)
        imglist.sort()
        for img in imglist:
            if img[-3:] == 'png':
                index = int(img[0:5])
#                 print(index,img)
                if( index<=64999 and img == dataset[index][0]):#training data from 00000.png to 64999.png
                    shutil.copy(path+file+'/'+img, output_path+'train/'+img)
                elif( index>64999 and index<=69999):#testing data from 65000.png to 69999.png
                    shutil.copy(path+file+'/'+img, output_path+'test/'+img)
```
